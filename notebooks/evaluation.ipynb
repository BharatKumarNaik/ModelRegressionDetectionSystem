```python
import pytest
from deepeval import evaluate
from deepeval.metrics import AnswerRelevancyMetric
from deepeval.test_case import LLMTestCase

def test_ans_relevance():
    input_text="what is the capital of France?"
    actual_output = "The capital of France is Paris."

    testcase = LLMTestCase(
        input = input_text,
        actual_output= actual_output
    )

    # Define Metrics
    metric = AnswerRelevancyMetric(threshold =0.7)

    # Run Evaluation
    metric.measure(testcase)

    print(metric.score)
    print(metric.success)
    print(metric.reason)
```

```python
import sqlglot

def evaluate_sql_structure(generated_sql: str, golden_sql: str) -> bool:
    try:
        # 1. Parse both queries into sqlglot AST objects
        # Specify the dialect of your SQL server (e.g., 'tsql' for Microsoft SQL Server)
        gen_ast = sqlglot.parse_one(generated_sql, read="tsql")
        gold_ast = sqlglot.parse_one(golden_sql, read="tsql")
        
        # 2. Transpile them into a standardized, clean string format
        # This strips out random white spaces, uniformizes keywords, and standardizes structure
        normalized_gen = gen_ast.sql(dialect="tsql", pretty=False)
        normalized_gold = gold_ast.sql(dialect="tsql", pretty=False)
        
        # 3. Compare the clean, standardized versions
        return normalized_gen == normalized_gold
        
    except sqlglot.errors.ParseError as e:
        # If the LLM generated invalid SQL syntax, the AST parser will fail immediately.
        # This is a huge win for regression testing—fuzzy matching can't catch syntax errors!
        print(f"SQL Syntax Error detected: {e}")
        return False

# --- TEST CASE ---
llm_query = "select u.name from users as u join orders o on u.id=o.user_id where o.status = 'Paid'"
gold_query = "SELECT users.name FROM users INNER JOIN orders ON users.id = orders.user_id WHERE orders.status = 'Paid';"

is_match = evaluate_sql_structure(llm_query, gold_query)
print(f"Do they structurally match? {is_match}") 
# Output: True
```

```python
from sqlglot import exp

parsed = sqlglot.parse_one("SELECT name FROM users WHERE age > 30", read="tsql")

# Find all table names in the query
tables = [table.name for table in parsed.find_all(exp.Table)]
print(tables)  # Output: ['users']

# Find all columns used in the WHERE condition
columns = [column.name for column in parsed.find(exp.Where).find_all(exp.Column)]
print(columns)  # Output: ['age']
```
