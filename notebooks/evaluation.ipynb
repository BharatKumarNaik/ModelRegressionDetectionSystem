```python
files = Path(TEST_ARTIFACTS_DIR).glob("*.json")
latest = max(files, key=lambda f: f.stat().st_mtime)
print(latest)
```

```python
latest_test_artifact = JsonUtils.read_json(latest)
```

```python
print(latest_test_artifact[0].keys())
print(golden_dataset[0].keys())
```

```python
from thefuzz import fuzz
# Check the similarity score
for i in range(len(latest_test_artifact)):
    g_sql_script = golden_dataset[i]['sql_script']
    t_sql_script = latest_test_artifact[i]['sql_script']
    # Compares the entire strings character-by-character (0-100); 100 means both strings are identical.
    print(f"Similarity score using ratio: {fuzz.ratio(g_sql_script, t_sql_script)}")

    # Finds the best matching substring within the longer string (0-100); useful when one string contains the other.
    print(f"Similarity score using partial ratio: {fuzz.partial_ratio(g_sql_script, t_sql_script)}")

    # Ignores word order by sorting tokens before comparison (0-100); useful when the same SQL clauses appear in a different order.
    print(f"Token sort ratio similarity score: {fuzz.token_sort_ratio(g_sql_script, t_sql_script)}")

    # Ignores duplicate/common tokens and compares the unique token sets (0-100); useful when one SQL query contains extra conditions or columns.
    print(f"Token sort ratio similarity score: {fuzz.token_set_ratio(g_sql_script, t_sql_script)}")
    print("--"*30)
```
